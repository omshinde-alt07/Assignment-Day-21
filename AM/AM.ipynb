1. Matrix Operations using NumPy

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)

print(result)

print(A.T)

det = np.linalg.det(A)
print(det)




```

```python
A = np.array([[2, 1],
              [1, 1]])

B = np.array([5, 3])

solution = np.linalg.solve(A, B)
print(solution)
```

```python
import time

arr = list(range(1000000))

start = time.time()

total = 0
for i in arr:
    total += i

end = time.time()

print("Loop Time:", end - start)
```

```python
arr_np = np.arange(1000000)

start = time.time()

total = np.sum(arr_np)

end = time.time()

print("NumPy Time:", end - start)
```

NumPy is much faster than Python loops
Loop takes more time because it processes element by element
NumPy uses vectorized operations and optimized C code internally

NumPy broadcasting is a technique that allows us to perform operations on arrays of different shapes without manually resizing them.
In simple words, NumPy automatically expands the smaller array so that both arrays have compatible shapes.

Why it is useful:
- No need to write loops
- Code becomes shorter and cleaner
- Faster execution
- Saves memory

Q.3
Vectorisation and loops are two different ways of performing operations on data. 
In loops, elements are processed one by one using a for loop, which makes the execution slower because Python handles each iteration separately. 
On the other hand, vectorisation operates on the entire array at once without using explicit loops, making the code shorter and more efficient. NumPy is faster because it uses optimized C code internally, avoids the overhead of Python loops, stores data in contiguous memory for better performance, and can perform operations in parallel.
