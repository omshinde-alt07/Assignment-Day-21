```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate data
np.random.seed(0)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.randn(100) * 0.2

degrees = [1, 2, 5]
errors = []

plt.scatter(X, y, label="Data")

for d in degrees:
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    y_pred = model.predict(X_poly)
    error = mean_squared_error(y, y_pred)
    errors.append(error)
    
    plt.plot(X, y_pred, label=f"Degree {d}")

plt.legend()
plt.title("Underfitting vs Overfitting")
plt.show()
```

Q.2

```python
plt.plot(degrees, errors, marker='o')
plt.xlabel("Model Complexity (Degree)")
plt.ylabel("Training Error (MSE)")
plt.title("Error vs Complexity")
plt.show()
```

Q.3

Bias is the error that occurs when a model is too simple to capture the underlying pattern in the data. Because of this, the model makes strong assumptions and fails to learn important relationships, leading to underfitting. For example, fitting a straight line to data that actually follows a curve results in high bias.
Variance, on the other hand, occurs when a model is too complex and tries to learn every detail in the data, including noise. This makes the model highly sensitive to small changes in the dataset and leads to overfitting. For example, a very wavy curve that passes through every data point represents high variance.
The optimal model lies between these two extremes. It is neither too simple nor too complex and is able to generalize well to new data. In this case, a model with degree 2 is considered optimal because it balances both bias and variance effectively.

Q1 — Difference between Regression and Classification

Regression and classification are types of supervised learning. Regression is used when the target variable is continuous, meaning it can take any numerical value. For example, predicting house prices or temperature are regression problems. Classification is used when the target variable is categorical, meaning it belongs to specific classes. For example, spam detection (spam or not spam) or predicting pass/fail are classification problems.

Q2 — Coding: calculate_mse(y_true, y_pred)
def calculate_mse(y_true, y_pred):
    return sum((y_true - y_pred) ** 2) / len(y_true)

This function calculates Mean Squared Error by finding the average of squared differences between actual and predicted values

Q3 — Bias–Variance Tradeoff

The bias–variance tradeoff explains the balance between a model being too simple and too complex. Bias is the error caused by a simple model that cannot capture the data patterns properly, leading to underfitting. Variance is the error caused by a complex model that learns noise along with data, leading to overfitting.
In underfitting, the model has high bias and performs poorly on both training and test data. In overfitting, the model has high variance and performs well on training data but poorly on new data. The goal is to find a model that balances both bias and variance for better performance.
